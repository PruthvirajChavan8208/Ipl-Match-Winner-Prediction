```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv("new_cleaned_ipl.csv")
```

```python
ds.info()
```

```python
ds=ds.drop(columns=['Unnamed: 0'])
```

```python
ds.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
ds['team1'].unique()
```

```python
all_teams=pd.concat([ds['team1'],ds['team2'],ds['toss_winner'],ds['winner']]).unique()
le.fit(all_teams)
```

```python
ds['team1']=le.transform(ds['team1'])
ds['team2']=le.transform(ds['team2'])
ds['toss_winner']=le.transform(ds['toss_winner'])
ds['winner']=le.transform(ds['winner'])
```

```python
ds['team1'].unique()
```

```python
cle=LabelEncoder()
```

```python
ds['city'].unique()
```

```python
ds['city']=cle.fit_transform(ds['city'])
```

```python
ds['city'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
tle=LabelEncoder()
```

```python
tle.fit(ds['toss_decision'])
```

```python
ds['toss_decision']=tle.transform(ds['toss_decision']) #bat=0.field=1
```

```python
x=ds.drop(columns=['winner'])
y=ds['winner']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('x_train',x_train.shape)
print('y_train',y_train.shape)
print('x_test',x_test.shape)
print('y_test',y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(n_estimators=100,max_depth=8,min_samples_split=5,random_state=42)
```

```python
RFC.fit(x_train,y_train)
```

```python
RFC.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report
y_pred=RFC.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy of Model is:{accuracy*100:.2f}%")
print("\n Classification report:\n",classification_report(y_test,y_pred))
```
